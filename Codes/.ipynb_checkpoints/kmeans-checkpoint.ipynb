```python
epss=np.linspace(1, 20, 51)
score=list()
for eps in epss:
    dbs=DBSCAN(eps=eps)
    dbs.fit(X)
    c = Counter()
    c = Counter(dbs.labels_)
    if len(set(dbs.labels_))>1:
        s=silhouette_score(X, dbs.labels_)
    else:
        s=None
    score.append(s)
    print (s,c)
#plt.plot(epss, score)
```

segmentation based on diabetes prevalence or obesity prevalence

```python
df_levels=pd.DataFrame()
df_levels['level_diabetes']=pd.qcut(df['prevalence of diabetes'], [0, .2, .4, .6, .8, 1.], labels=False)
df_levels['level_obesity']=pd.qcut(df['prevalence of obesity'], [0, .2, .4, .6, .8, 1.], labels=False)

df_levels.index=df.index
```

```python
set(df_levels['level_diabetes'])
set(df_levels['level_obesity'])
```

```python
#most diabetes prevalent counties
```

```python
df_subset_1=df[df_levels['level_diabetes']==4][columns_subset]
normed_data_subset_1=preprocessing.scale(df_subset_1)
normed_df_subset_1=pd.DataFrame(normed_data_subset_1, index=df_subset_1.index, columns=df_subset_1.columns)
```

```python
from sklearn.metrics import silhouette_score
X=normed_df_subset_1.as_matrix()
k_state_scores=list()
for i in range(2,21):
    for state in [0, 10, 50]:
        kmeans = KMeans(n_clusters=i, random_state=state).fit(X)
        c = Counter(kmeans.labels_)
        s=silhouette_score(X, kmeans.labels_)
        k_state_scores.append((s,i,state))
k_state_scores=sorted(k_state_scores, key=lambda x: x[0])
print (k_state_scores[-1])
    #print (i,s,c) 
```

```python
X=normed_df_subset_1.as_matrix()
kmeans = KMeans(n_clusters=2, random_state=50).fit(X)
```

```python
normed_df_subset_1['group']=kmeans.labels_
```

```python
for c in normed_df_subset_1.columns:
    if c!='group':
        plt.figure()
        sns.boxplot(x='group', y=c, data=normed_df_subset_1)
        plt.title(c)
        plt.show()
```
