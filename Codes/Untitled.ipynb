```r
Map<- function (df,var){
        df_map<-df[,c("FIPS",var)]
        print (df_map[var])
        df_map<-na.omit(df_map)

        colnames(df_map)<-c("region","value")
        county_choropleth(df_map, title =var, legend='')
}
```
