```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

preprocessing data

```python
#read the complete dataset
url='https://raw.githubusercontent.com/cathyxinxyz/Capstone_Project_1/master/Datasets/Combined_data.csv'
df=pd.read_csv(url,index_col='FIPS',encoding="ISO-8859-1")
```

```python
#standardize data: (x-mean(x))/std
from sklearn import preprocessing
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype

scaled_columns=[c for c in df.columns if is_numeric_dtype(df[c])]
df_for_scale=df[scaled_columns].dropna()
X=df_for_scale.values
scaler=  preprocessing.StandardScaler()
X_scaled = scaler.fit_transform(X)
scaled_df = pd.DataFrame(X_scaled, index=df_for_scale.index, columns=df_for_scale.columns)
```

```python
scaled_df.shape
```

```python
scaled_df.info()
```

```python
#replace space in variable names with _
scaled_df.rename(columns=lambda x:'above'.join(x.split('>=')), inplace=True)
scaled_df.rename(columns=lambda x:'under'.join(x.split('<')), inplace=True)
scaled_df.rename(columns=lambda x:'_'.join(x.split(' ')), inplace=True)
```

```python
#construct the first linear regression model
vars=list(scaled_df.columns.difference(['prevalence_of_diabetes',
                                             'prevalence_of_obesity',
                                             'Adult_db', 
                                             'Adult_ob']))
features='+'.join(vars)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, scaled_df, return_type='dataframe')

mod = smf.OLS(y, X)
result=mod.fit()

#Y = scaled_df['prevalence of diabetes']
#X = scaled_df[scaled_df.columns.difference(['prevalence of diabetes',
                                             #'prevalence of obesity',
                                            # 'Adult_db', 
                                             #'Adult_ob'])]
#X = sm.add_constant(X)
#model = sm.OLS(Y,X)
```

```python

#the results incidate a strong multicollinearity among variables that quantify 
#the race/ethnicity composition: White, Black, Hispanic and American_Indian_or_Alaska_Native
#As indicated in the section of exploratory data analysis, the percentage of White people in the population
# negatively correlates with percentage of Black people, Hispanic people, or American_Indian_or_Alaska_Native people

#In addition, there is also moderate multicollinearity among variables Low_Access_Overall,
#Low_Access_Child, Low_Access_Seniors, Low_Access_Overall, Low_Access_low_income, as indicated in exploratory analysis
#there are positive correlations among these variables 

#lastly, low_security and very_low_security have a moderate level of 
#multicollinearity, this is also 


#In order to remove the variables with high multicollinearity, I remove variables Black, Hispanic and American_Indian_or_Alaska_Native, Asian
#and Low_Access_Child, Low_Access_Seniors, 

vars=list(set(vars).difference(set(['Black','Hispanic','Asian','American_Indian_or_Alaska_Native',
                                    'Low_Access_Child','Low_Access_Seniors','Low_Access_low_income',
                                    'Very_low_insecurity'])))
```

```python
features='+'.join(vars)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, scaled_df, return_type='dataframe')

mod = sm.OLS(y, X)
result=mod.fit()
```

```python
result.summary()
```

After removing the variables that have collinearity with other variables, R square decreases from 0.728 to 0.671, which is reasonable since I only use a subset of the variables in the model. In addition, coefficients of several variables become statistically significant (p value falls under 0.05). Such variables include Low_Access_Overall, Hawaiian_or_Pacific_Islander and White, which confirms the impact of multicollinearity. However, p value of coefficient of variable 'under18' increases to above 0.05, suggeting that effect of this variable on the dependent variable becomes insignificant. This can be because of the increase in the effect of other variables on the dependent variable. 

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

```python
vif
```

vif test shows that there is no sever multicollinearity this time

The regression results indicate that there are several variables whose coefficients are not statistically significant (p value >0.05). Therefore, I remove the insignificant independent variables ('under18', 'Specialized' and 'Supercenter')

```python
vars=list(set(vars).difference(set(['under18', 'Specialized','Supercenter','Fast_food'])))
```

```python
features='+'.join(vars)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, scaled_df, return_type='dataframe')

mod = smf.OLS(y, X)
result=mod.fit()
```

```python
result.summary()
```

after I removed the variables with insignificant coefficients, R square does not change much. However, this may not be a complete model without addressing the possible interactions. For example, the impact of food environment and physical activity on prevalence of diabetes may differ by the percentage of population above 65 years old, since the risk of type II diabetes increases as age increases. In order to detect all the possible interactions, I added all pairs of interactions into the dataset. 

```python
from itertools import combinations
vars_with_interaction=list()
vars_with_interaction.extend(vars)
for var_pair in combinations(vars,2):
    scaled_df[var_pair[0]+'_'+var_pair[1]]=scaled_df[var_pair[0]]*scaled_df[var_pair[1]]
    vars_with_interaction.append(var_pair[0]+'_'+var_pair[1])
    #print (var_pair)
```

```python
features='+'.join(vars_with_interaction)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, scaled_df, return_type='dataframe')

mod = smf.OLS(y, X)
result=mod.fit()
```

```python
result.summary()
```

After including all the possible interaction terms, there is an interesting phenomenon: main effects that are used to be statiscially significant become insignificant. These variables include: 'Low_Access_Overall', 'Hawaiian_or_Pacific_Islander', 'Farm', 'Recreation_facility' and 'Convenience'. This may suggest that effects of these variables are stronger given the existence of other variables. For example, variable 'Farm' has statiscially insignificant effect on prevalence of diabetes but there is a significant effect by the interaction between 'Farm' and 'White', which may be because that available farms will only affect prevalence of diabetes if some people are White. However, this does not make sense since there are nearly no states with no White residents. Therefore, this may indicate a strong dependence of impact of available Farms on the percentage of White population. In order to more clearly observe the regression results, we remove the interaction terms with statistically insignificant coefficients. 

```python
alpha = 0.05
#result.summary().tables[1].data[1:]
param_table= pd.DataFrame(data = [x for x in result.summary().tables[1].data[1:] if (float(x[4]) < alpha) | (x[0] in vars)], columns = result.summary().tables[1].data[0])
```

```python
param_table
```

```python
vars_to_keep=list(param_table[''])
features='+'.join(vars_to_keep)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, scaled_df, return_type='dataframe')

mod = smf.OLS(y, X)
result=mod.fit()
```

```python
result.summary()
```

```python
#further remove two interaction terms with insignificant coefficient:
# Hawaiian_or_Pacific_Islander_Poverty_rate and Low_insecurity_Poverty_rate
vars_to_keep=list(set(vars_to_keep).difference(set(['Hawaiian_or_Pacific_Islander_Poverty_rate','Low_insecurity_Poverty_rate'])))
features='+'.join(vars_to_keep)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, scaled_df, return_type='dataframe')

mod = smf.OLS(y, X)
result=mod.fit()
```

```python
result.summary()
```

```python
#outliers and high leverage points
fig, ax = plt.subplots(figsize=(24,16))
fig=sm.graphics.influence_plot(result, ax= ax)
plt.show()
```

Potential outliers: county with FIPS code 1085,46121,2290 and 12041
potential high leverage points: county with FIPS 15001, 15009, 8053, 41055

```python
#remove each observation one by one and see the difference in model fitting results

Outliers=[1085,46121,2290,12041]
High_levg=[15001, 15009, 8053, 41055]
results_outliers=dict()
for ol in Outliers:
    df_dropdata=scaled_df.drop([ol])
    y, X = dmatrices('prevalence_of_diabetes ~1+' + features, df_dropdata, return_type='dataframe')

    mod = smf.OLS(y, X)
    results_outliers[ol]=mod.fit()

results_high_levg=dict()
for hl in High_levg:
    df_dropdata=scaled_df.drop([hl])
    y, X = dmatrices('prevalence_of_diabetes ~1+' + features, df_dropdata, return_type='dataframe')

    mod = smf.OLS(y, X)
    results_high_levg[hl]=mod.fit()

Outliers.extend(High_levg)
df_dropall=scaled_df.drop(Outliers)
y, X = dmatrices('prevalence_of_diabetes ~1+' + features, df_dropall, return_type='dataframe')
mod = smf.OLS(y, X)
results_drop_all=mod.fit()
```

```python
#results_outliers[1085].summary()
#results_high_levg[8053].summary()
results_drop_all.summary()
```

R square does not change much (0.721 to 0.725) even after all potential outliers and high leverage points are dropped. There is no dramatic change in coefficient estimates either.
